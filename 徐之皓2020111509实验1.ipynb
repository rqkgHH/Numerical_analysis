## 第一问

```python
import math
a = 1
b = 2
s = 0
def fun1(x):
    return 2/3*(x**3)*(math.e**(x**2))

def fhtx(): ##复合梯形公式
    n = 100000
    while True:
        n += 1000
        h = (b-a) / n
        sum1 = 0
        for k in range(1,n):
            ai = a + k * h
            sum1 = sum1 + fun1(ai)

        T = h / 2 * (fun1(a) + fun1(b) + 2 * sum1) #近似值计算
        
        if abs(T - math.e**4) <= 0.5*10**(-8):
            break ##记录循环次数
        else:
            print(T - math.e**4)
            continue
    return T,n

def fhsim(): ##复合辛普森公式
    n = 100
    while True:
        n += 1
        h = (b-a) / n
        sum1 = 0
        sum2 = 0
        for k in range(1,n+1):
            ai = a + k * h
            sum1 = sum1 + fun1(ai)
            bi = ai - 1/2* h
            sum2 = sum2 + fun1(bi)
        
        T = h / 6 * (fun1(a) + fun1(b) + 2 * sum1 + 4 * sum2 - 2 * fun1(b)) #近似值计算
        
        if abs(T - math.e**4) <= 0.5*10**(-8):
            break ##记录循环次数
        else:
            print(T - math.e**4)
            continue
    return T,n
```

```python
def romberg():
    n = 0
    a = 1
    b = 2
    R = [[0] * 10000] * 10000
    # R = [[0]*(n+1) for_in range(n+1)]
    h = b-a
    R[0][0] = h/2+(fun1(a)+fun1(b))
    while True:
        n = n+1
        for i in range(1,n+1):
            h = h/2
            sum3 = 0
            for k in range(1,2**i,2):
                sum3 += fun1(a+k*h)
            R[i][0] = R[i-1][0]/2 + sum3*h
            for j in range(1,i+1):
                R[i][j] = (4**j*R[i][j-1]-R[i-1][j-1])/(4**j-1)
        if abs(R[n][n]) <= 0.5*10**(-8): 
            break
        else:
            print(R[n][n])
            continue
            
    return R[n][n]+math.e**4,i
```
